#Первая часть Python

##Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
df_ratings = pd.read_csv("/content/u.data.csv", sep="\t", names=['user_id', 'item_id', 'rating',	'timestamp'])
df_ratings
```

```python
df_movies = pd.read_csv("/content/u.item.csv", sep="|", encoding='latin-1', names=['id', 'title', 'date', 'Empty', 'link', 'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])
```

```python
df_movies = df_movies.drop('Empty', axis=1)
```

##Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
top_one_user = df_ratings.groupby('user_id').count().sort_values('rating', ascending=False).reset_index().iloc[0]['user_id']
top_one_user
```

##Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь

```python
df_ratings_405 = df_ratings[df_ratings['user_id'] == top_one_user]
df_ratings_405
```

##Добавьте к датафрейму из задания 3 столбцы:
По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет


```python
df_genres = df_movies.drop(['title', 'link'], axis=1)
df_genres
```

```python
df_ratings_405 = df_ratings_405.merge(df_genres, how='left', left_on='item_id', right_on='id')
```

```python
df_ratings_405 = df_ratings_405.drop('id', axis=1)
df_ratings_405
```

```python
df_ratings_405['date'] = df_ratings_405['date'].apply(lambda x: x.split('-')[-1])
df_ratings_405
```

##столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей


```python
df_all_ratings = df_ratings.groupby('item_id')['rating'].count().reset_index()
df_all_ratings = df_all_ratings.rename(columns={'rating': 'all_ratings'})
df_all_ratings
```

```python
df_sum_ratings = df_ratings.groupby('item_id')['rating'].sum().reset_index()
df_sum_ratings = df_sum_ratings.rename(columns={'rating': 'sum_ratings'})
df_sum_ratings
```

```python
df_ratings_405 = df_ratings_405.merge(df_all_ratings, how='left')
df_ratings_405 = df_ratings_405.merge(df_sum_ratings, how='left')
```

```python
df_ratings_405
```

```python
df_for_ml = df_ratings_405.drop(['user_id', 'item_id', 'timestamp'], axis=1)
#приведу колонку date  к типу int32
df_for_ml['date'] = df_for_ml['date'].astype('int32')
```

##Сформируйте X_train, X_test, y_train, y_test

```python
from sklearn.model_selection import train_test_split
```

```python
df_for_ml.columns
```

```python
X,y = df_for_ml[['date', 'unknown', 'Action', 'Adventure', 'Animation',
       "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
       'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi',
       'Thriller', 'War', 'Western', 'all_ratings', 'sum_ratings']], df_for_ml[['rating']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

##Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

##Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_predict)
```

##Средствами спарка вывести среднюю оценку для каждого фильма

```python
df_rating.createOrReplaceTempView('rating')
```

```python
df_count_sum = spark.sql('SELECT _c1, count(_c2), sum(_c2), sum(_c2) / count(_c2) as avg_rat  FROM rating GROUP BY _c1')
df_count_sum.show()
```

##Посчитайте средствами спарка среднюю оценку для каждого жанра

```python
df_movie.createOrReplaceTempView('movie')
```

```python
df_movie.show()
```

```python
df_rating.show()
```

```python
df_rating = df_rating.selectExpr("_c0 as user_id", "_c1 as item_id" , "_c2 as rating", "_c3 as timestamp")
df_rating.show()
```

```python
df_join = df_rating.join(df_movie, df_rating.item_id == df_movie._c0, "inner")
df_join.show()
```

```python
list_of_genres = ['_c5','_c6','_c7','_c8','_c9','_c10','_c11','_c12','_c13','_c14','_c15','_c16','_c17','_c18','_c19','_c20','_c21','_c22','_c23']
def avg_of_genre(list_of_genres):
  for genre in list_of_genres:
    df_filter = df_join.where(df_join[genre] == 1)
    df_filter.createOrReplaceTempView('join_filter')
    df_filter = spark.sql(f"SELECT sum(rating) / count(rating) as avg_of_genre{genre} FROM join_filter GROUP BY {genre}")
    df_filter.show()
avg_of_genre(list_of_genres)
```

##В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами (по количеству оценок, либо по самой оценке - на Ваш выбор)

```python
# по количеству оценок
most_popular = spark.sql('SELECT _c1, count(_c2), sum(_c2), sum(_c2) / count(_c2) as avg_rat  FROM rating GROUP BY _c1 order by count(_c2) DESC limit 5')
```

```python
most_popular.show()
```

```python
most_unpopular = spark.sql('SELECT _c1, count(_c2), sum(_c2), sum(_c2) / count(_c2) as avg_rat  FROM rating GROUP BY _c1 order by count(_c2) limit 5')
most_unpopular.show()
```
